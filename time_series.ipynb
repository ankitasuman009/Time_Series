```python
data_cols = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
#resample to weekly frequency and aggregate with mean
weekly_mean = df[data_cols].resample('W').mean()
weekly_mean.head(3)
```

```python
#weekly_mean has data 1/7 of daily time series.
print(df.shape[0])
print(weekly_mean.shape[0])
```

```python
# plot the daily and weekly Solar time series together over a single six-month period to compare them.
start, end = '2017-01', '2017-06'
fig, ax = plt.pyplot.subplots()
ax.plot(df.loc[start:end, 'Solar'],
       marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(weekly_mean.loc[start:end, 'Solar'],
       marker='o', markersize=8, linestyle='-',
       label='Weekly Mean Resample')
ax.set_ylabel('Solar Production (GWh)')
ax.legend()
```

```python
#Compute the monthly sums
monthly_sum = df[data_cols].resample('M').sum(min_count=28)
monthly_sum.head(3)
```

```python
#monthly time series by plotting the electricity consumption as a line plot

fig, ax = plt.pyplot.subplots()
ax.plot(monthly_sum['Consumption'], color='black', label='Consumption')

#wind and solar power production together as a stacked area plot

monthly_sum[['Wind', 'Solar']].plot.area(ax=ax, linewidth=0)
ax.xaxis.set_major_locator(plt.dates.YearLocator())
ax.legend()
ax.set_ylabel('Monthly Total (GWh)')
```

```python
#Compute the annual sums
annual_sum = df[data_cols].resample('A').sum(min_count=360)

# The default index of the resampled DataFrame is the last day of each year, so to make life easier, set the index
# to the year component

annual_sum = annual_sum.set_index(annual_sum.index.year)
annual_sum.index.name = 'Year'

# Compute the ratio of Wind+Solar to Consumption

annual_sum['Wind+Solar/Consumption'] = annual_sum['Wind+Solar'] / annual_sum['Consumption']
annual_sum.tail(3)
```

```python
#plot the wind + solar share of annual electricity consumption as a bar chart.
ax = annual_sum.loc[2012:, 'Wind+Solar/Consumption'].plot.bar(color='C0')
ax.set_ylabel('Fraction')
ax.set_ylim(0, 0.3)
ax.set_title('Wind+Solar Annual Electricity Consumption')
plt.pyplot.xticks(rotation=0)
```

```python
# Compute the centered 7-day rolling mean

df_7d = df[data_cols].rolling(7, center=True).mean()
df_7d.head(5)
```

```python
#2006-01-04 is the midpoint of the first rolling window.
start, end = '2017-01', '2017-06'

## Plot daily, weekly resampled, and 7-day rolling mean time series together

fig, ax = plt.pyplot.subplots()
ax.plot(df.loc[start:end, 'Solar'],
       marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(weekly_mean.loc[start:end, 'Solar'],
       marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.plot(df_7d.loc[start:end, 'Solar'],
       marker='.', linestyle='-', label='7-D Rolling Mean')
ax.set_ylabel('Solar Production )GWh')
ax.legend()
```

```python
#compute the 365-day rolling mean of our OPSD data
df_365d = df[data_cols].rolling(window=365, center=True, min_periods=360).mean()
```

```python
# Plot daily, 7-day rolling mean, and 365-day rolling mean time series
fig, ax = plt.pyplot.subplots()
ax.plot(df['Consumption'], marker='.', markersize=2, color='0.6',
       linestyle='None', label='Daily')
ax.plot(df_7d['Consumption'], linewidth=2, label='Trend(7-D Rolling Mean)')
ax.plot(df_365d['Consumption'], color='0.2', linewidth=3,
       label='Trend(365-D Rolling Mean)')
ax.xaxis.set_major_locator(plt.dates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption')
```

```python
# Plot 365-day rolling mean time series of wind and solar power
```

```python
fig, ax = plt.pyplot.subplots()
for col in ['Wind', 'Solar', 'Wind+Solar']:
    ax.plot(df_365d[col], label=col)
    ax.xaxis.set_major_locator(plt.dates.YearLocator())
    ax.set_ylim(0, 400)
    ax.legend()
    ax.set_ylabel('Production (GWh)')
    ax.set_title('Trends in Electricity Production (365-D Rolling Means)')
```
